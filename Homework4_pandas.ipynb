# Compare The Trend Of Crypto-Currency

## import packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import requests
import io
```

## cookies and crumb

```python
session = requests.Session()
crumbText = session.get("https://finance.yahoo.com/quote").text
crumbIndex = crumbText.find('CrumbStore":{"crumb":"') +22
crumb = crumbText[crumbIndex : crumbIndex + 11]
```

## some strings and variables

```python
download_str = "https://query1.finance.yahoo.com/v7/finance/download/"
period_str = "?period1=1491321600&period2=1554393600&interval=1d&events=history&crumb="

BTC_USD = session.get(download_str + "BTC-USD" + period_str + crumb)
ETH_USD = session.get(download_str + "ETH-USD" + period_str + crumb)
BNB_USD = session.get(download_str + "BNB-USD" + period_str + crumb)
LTC_USD = session.get(download_str + "LTC-USD" + period_str + crumb)
XRP_USD = session.get(download_str + "XRP-USD" + period_str + crumb)
ZEC_USD = session.get(download_str + "ZEC-USD" + period_str + crumb)
```

## read csv

```python
df_BTC = pd.read_csv(io.StringIO(BTC_USD.text))
df_ETH = pd.read_csv(io.StringIO(ETH_USD.text))
df_BNB = pd.read_csv(io.StringIO(BNB_USD.text))
df_LTC = pd.read_csv(io.StringIO(LTC_USD.text))
df_XRP = pd.read_csv(io.StringIO(XRP_USD.text))
df_ZEC = pd.read_csv(io.StringIO(ZEC_USD.text))
```

## divide the exchage rates by the initial exchange rates

```python
plt.plot(df_BTC["Adj Close"] / df_BTC["Adj Close"][0], label = "BTC")
plt.plot(df_ETH["Adj Close"] / df_ETH["Adj Close"][0], label = "ETH")
plt.plot(df_BNB["Adj Close"] / df_BNB["Adj Close"][0], label = "BNB")
plt.plot(df_LTC["Adj Close"] / df_LTC["Adj Close"][0], label = "LTC")
plt.plot(df_XRP["Adj Close"] / df_XRP["Adj Close"][0], label = "XRP")
plt.plot(df_ZEC["Adj Close"] / df_ZEC["Adj Close"][0], label = "ZEC")

plt.legend(prop={'size': 25})

fig = plt.gcf()
fig.set_size_inches(16, 9)
```
