# 擊敗老師的神經網路 - 比較相同數量參數下的訓練效果

There about 240,000 parameters each
- 1 hidden layer of 300 neurons  

- 2 hidden layers of 233 neurons each  

- 3 hidden layers of 200 neurons each  

- 4 hidden layers of 179 neurons each  

- 5 hidden layers of 164 neurons each  


## import packages

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
from keras.datasets import mnist
from keras.utils import np_utils
```

## reshape data

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)

## 1-hot encoding
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)
```

## build a neural network

### 1 hidden layer of 300 neurons  

```python
model1 = Sequential()
# hidden layers
model1.add(Dense(units = 300, input_dim = 784, activation = 'sigmoid'))
# outpput layer
model1.add(Dense(units = 10, activation = 'softmax'))
model1.compile(loss = 'mse', optimizer = SGD(lr = 0.09487), metrics = ['accuracy'])
model1.summary()
model1.fit(x_train, y_train, batch_size = 100, epochs = 20)
```

### 2 hidden layers of 233 neurons each  

```python
model2 = Sequential()
# hidden layers
model2.add(Dense(units = 233, input_dim = 784, activation = 'sigmoid'))
model2.add(Dense(units = 233, activation = 'sigmoid'))
# outpput layer
model2.add(Dense(units = 10, activation = 'softmax'))
model2.compile(loss = 'mse', optimizer = SGD(lr = 0.09487), metrics = ['accuracy'])
model2.summary()
model2.fit(x_train, y_train, batch_size = 100, epochs = 40)
```

### 3 hidden layers of 200 neurons each

```python
model3 = Sequential()
# hidden layers
model3.add(Dense(units = 200, input_dim = 784, activation = 'sigmoid'))
model3.add(Dense(units = 200, activation = 'sigmoid'))
model3.add(Dense(units = 200, activation = 'sigmoid'))
# outpput layer
model3.add(Dense(units = 10, activation = 'softmax'))
model3.compile(loss = 'mse', optimizer = SGD(lr = 0.09487), metrics = ['accuracy'])
model3.summary()
model3.fit(x_train, y_train, batch_size = 100, epochs = 70)
```

###  4 hidden layers of 179 neurons each

```python
model4 = Sequential()
# hidden layers
model4.add(Dense(units = 179, input_dim = 784, activation = 'sigmoid'))
model4.add(Dense(units = 179, activation = 'sigmoid'))
model4.add(Dense(units = 179, activation = 'sigmoid'))
model4.add(Dense(units = 179, activation = 'sigmoid'))
# outpput layer
model4.add(Dense(units = 10, activation = 'softmax'))
model4.compile(loss = 'mse', optimizer = SGD(lr = 0.09487), metrics = ['accuracy'])
model4.summary()
model4.fit(x_train, y_train, batch_size = 100, epochs = 100)
```

### 5 hidden layers of 164 neurons each

```python
model5 = Sequential()
# hidden layers
model5.add(Dense(units = 164, input_dim = 784, activation = 'sigmoid'))
model5.add(Dense(units = 164, activation = 'sigmoid'))
model5.add(Dense(units = 164, activation = 'sigmoid'))
model5.add(Dense(units = 164, activation = 'sigmoid'))
model5.add(Dense(units = 164, activation = 'sigmoid'))
# outpput layer
model5.add(Dense(units = 10, activation = 'softmax'))
model5.compile(loss = 'mse', optimizer = SGD(lr = 0.09487), metrics = ['accuracy'])
model5.summary()
model5.fit(x_train, y_train, batch_size = 100, epochs = 200)
```

## Evaluate the models

```python
score1 = model1.evaluate(x_test, y_test)
score2 = model2.evaluate(x_test, y_test)
score3 = model3.evaluate(x_test, y_test)
score4 = model4.evaluate(x_test, y_test)
score5 = model5.evaluate(x_test, y_test)
```

```python
print("1 hidden layer of 300 neurons :\t\t", score1[1])
print("2 hidden layers of 233 neurons each :\t", score2[1])
print("2 hidden layers of 233 neurons each :\t", score3[1])
print("4 hidden layers of 179 neurons each :\t", score4[1])
print("5 hidden layers of 164 neurons each :\t", score5[1])
```

## Summary
Neural networks have very different performance even if they have same number of parameters.  
Accroding to above, here might be a question, Why deepest one performs worst? or, How to assign those parameters is optimal? 

By the way, 1 hidden layer of 10 neurons can achieve 90% accuracy.
